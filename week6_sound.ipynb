```python
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd
import numpy as np
%matplotlib notebook
from scipy.signal import lfilter
```

sr:1초에 숫자가 몇번 있느냐, freq:반복되는 횟수,

모든 소리는 sin wave의 합이다. 
1. 음의 높낮이:sin의 freq
2. 음의 강도(진폭) 

1과2는 독립적으로 연관

Phasor: sin wave를 만드는 장치

amp=1이면 -1부터1까지//sr는 1초에 개별숫자가 얼마나 들어있는지 정의//dur는 소리가 얼마동안 지속될건지, freq: sin wave가 1초에 반복되는 횟수(1초에 sin wave가 100번 그려짐)

```python
# parameter setting
amp = 1          # range [0.0, 1.0]
sr = 10000       # sampling rate, Hz
dur = 1       # in seconds
freq = 10000.0    # frequency, Hz
```

sin함수에 degree를 넣지말고 pi를 이용해서..

t는 시간: arrange 1부터 (sr*dur+1)

t의 element는 10000개의 sample

★여기서 셤 많이나온다!!!!!!!!!!!

```python
# generate time
t = np.arange(1, sr * dur+1)/sr
```

time과 각도값 합침

```python
# generate phase
theta = t * 2*np.pi * freq
```

t와 theta는 길이가 같음

```python
# generate signal by cosine-phasor
s = np.sin(theta)
```

여기서부터 시험에 내지는 않겠다---

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t[0:5000], s[0:5000], '.')
ax.set_xlabel('time(s)')
ax.set_ylabel('real')
```

c = np.exp(theta*1j)


```python
# generate signal by complex-phasor
c = np.exp(theta*1j)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(t[0:5], c.real[0:5], c.imag[0:5], '.')
ax.set_xlabel('time (s)')
ax.set_ylabel('real')
ax.set_zlabel('imag')
```
