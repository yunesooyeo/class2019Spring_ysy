```python
# description
# https://www.kaggle.com/c/boston-housing

# TensorFlow and tf.keras
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python

(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
```

relu: Dense-- Hidden Layer
regression인 경우 optimizer는 rmsprop으로, loss는 mse로
classification인 경우 optimizer는 adam, loss는 entrophy로

```python
model = Sequential()
model.add(Dense(10, input_shape=[13], activation = 'relu'))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

140) 첫번쨰 layer에 node가 13개, 두번째 layer에 node가 10개 --> 130개

```python
# TensorFlow and tf.keras
import tensorflow as tf

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt

(x_train, y_train),(x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0
```

y train에는 총 6만개는 feature이 들어있음
        x train에는 

```python
plt.figure(figsize=(100,100)) # size of figure
for i in range(100):
    plt.subplot(10,10,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i], cmap=plt.cm.binary)
plt.show()
```

```python
img_rows = 28
img_cols = 28

model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(img_rows, img_cols)),
  tf.keras.layers.Dense(512, activation='relu'),
  tf.keras.layers.Dropout(0),
  tf.keras.layers.Dense(10, activation ='softmax')
])
```

```python
# Take a look at the model summary
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
Adam=tf.keras.optimizer.Adam
model.compile(optimizer=Adam(lr=0.001), loss='sparse_categorical_corssentropy', metrics=['accuracy'])
```

classification의 optimizer는 Adam사용
learning rate(lr)을 조절하여 속도, 정확성 조정

```python
print(model.input_shape)
print(model.output_shape)
```

```python
model.fit(x_train, y_train, epochs=1, batch_size=32)
```

```python
model.evaluate(x_test, y_test)
```
